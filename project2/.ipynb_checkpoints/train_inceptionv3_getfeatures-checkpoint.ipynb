#### NOTEBOOK 1

This notebook fits a inceptionv3 neural net with the imagenet weights and extracts the features from the train images.
The features and labels are stored as h5py files.

```python
import keras
```

```python
# keras imports
from keras.applications.inception_v3 import InceptionV3, preprocess_input
from keras.preprocessing import image
from keras.models import Model
from keras.models import model_from_json
from keras.layers import Input
from PIL import Image
```

```python
# other imports
from sklearn.preprocessing import LabelEncoder
import numpy as np
import glob
import cv2
import h5py
import os
import json
import datetime
import time
import PIL
```

```python
# load the user configs
with open('room_recognition/conf/conf_inceptionv3_getfeatures.json') as f:    
  config = json.load(f)
```

```python
# config variables
model_name  = config["model"]
weights     = config["weights"]
train_path  = config["train_path"]
test_path  =config["test_path"]
train_features_path   = config["train_features_path"]
train_labels_path   = config["train_labels_path"]

seed =config["seed"]



```

```python
# start time
print ("[STATUS] start time - {}".format(datetime.datetime.now().strftime("%Y-%m-%d %H:%M")))
start = time.time()
```

```python
# create the pretrained models
base_model = InceptionV3(include_top=True, weights=weights, input_tensor=Input(shape=(299,299,3)))
model = Model(inputs=base_model.input, outputs=base_model.get_layer('avg_pool').output)
image_size = (299, 299)

print ("[INFO] successfully loaded base model and model...")
```

```python
# path to training dataset
train_labels = os.listdir(train_path)
```

```python
# encode the labels
print ("[INFO] encoding labels...")
le = LabelEncoder()
le.fit([tl for tl in train_labels])
```

```python
# variables to hold features and labels
features = []
labels   = []
```

```python
# loop over all the labels in the folder
count = 1
for i, label in enumerate(train_labels):
  cur_path = train_path + "/" + label
  count = 1
  for image_path in glob.glob(cur_path + "/*.jpg"):
    img = image.load_img(image_path, target_size=image_size)
    x = image.img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = preprocess_input(x)
    feature = model.predict(x)
    flat = feature.flatten()
    features.append(flat)
    labels.append(label)
    print ("[INFO] processed - " + str(count))
    count += 1
  print ("[INFO] completed label - " + label)
```

```python
# encode the labels using LabelEncoder
le = LabelEncoder()
le_labels = le.fit_transform(labels)
```

```python
# get the shape of training labels
print ("[STATUS] training labels: {}".format(le_labels))
print ("[STATUS] training labels shape: {}".format(le_labels.shape))
```

```python
# save features and labels
h5f_data = h5py.File(train_features_path, 'w')
h5f_data.create_dataset('dataset_1', data=np.array(features))

h5f_label = h5py.File(train_labels_path, 'w')
h5f_label.create_dataset('dataset_1', data=np.array(le_labels))

h5f_data.close()
h5f_label.close()
```

```python
# end time
end = time.time()
print ("[STATUS] end time - {}".format(datetime.datetime.now().strftime("%Y-%m-%d %H:%M")))
```
